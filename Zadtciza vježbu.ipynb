```python
import math
def opsegKruga (r):
    return 2*r*math.pi
def povrsinaKruga(r):
    return r*r*math.pi
def promijer(r):
    return r*2
def volumenKruga(r):
    raz =4/3
    r3=r*r*r
    pi=math.pi
    f=raz*r3*pi
    return f
def krug(r):
    print ("Opseg kruga : "+str(opsegKruga(r))+" cm")
    print ("Povrsina kruga : "+str(povrsinaKruga(r))+" cm*cm")
    print ("Promjer kruga : "+str(promijer(r))+" cm")
    print ("Volumen kruga : "+str(volumenKruga(r))+" cm*cm*cm")    
def povrsinaKvadrata (a):
    return a*a
def opsegkKvadrata(a):
    return a*4
def kvadrat(a):
    print ("opseg kvadrata :"+ str(opsegkKvadrata(a))+" cm")
    print ("površina kvadrata :"+ str(opsegkKvadrata(a))+" cm*cm")
def povrsinaTrokuta(a,b):
    p=(a*b)/2
    return p
def opsegTrokuta(a,b,c):
    return a+b+c
def hipozenuzaTrokuta(a,b):
    ck=(a*a)+(b*b)
    c= math.sqrt(ck)
    return c
def reversHypotenusa(a,c):
    cc=c*c
    aa=a*a
    return math.sqrt(cc-aa)    
def trokut(a,b,c):
    print ("Površina trokuta :"+str(povrsinaTrokuta(a,b))+" cm*cm")
    print ("Opseg trokuta :"+str(opsegTrokuta(a,b,c))+" cm")
    print ("Duljina hipotenuze :"+str(hipozenuzaTrokuta(a,b))+" cm")
def sinAlpha (a,c):
    return a/c
def cosAlpha (b,c):
    return b/c
def tanAlpha(a,b):
    return a/b
def cotAlpha(b,a):
    return b/a
def alpha(a,b,c):
    print("Sinus alpha :"+str(sinAlpha(a,c)))
    print("Cosinus alpha :"+str(cosAlpha(b,c)))
    print("Tangensinus alpha :"+str(tanAlpha(a,b)))
    print("Cotangens alpha :"+str(cotAlpha(b,a)))
def povrsinaPralelograma(a,v):
    return a*v
def volumenPrizmeIValjka (b,h):
    return b*h
def oplosjePiramide(b,p):
    return b+p
def volumenPiramideIliStisca (b,h):
    return (1/3)*b*h
def nejednakosti(a,b):
    if b > a:
         return  print(str(a)+" < "+str(b))
    elif a == b:
         return print(str(a)+" = "+str(b))
    else:
         return print(str(a)+" > "+str(b))
def nejednakostiPlus(a,b,c):
    if b > a:
        if c =='>':
            return print("Tocno")
        elif c=='<':
            return print("Tocno")
        elif c=='=':
            return print("Tocno")
        else:
            return print("Netocno")
    elif a == b:
        if c =='>':
            return print("Tocno")
        elif c=='<':
            return print("Tocno")
        elif c=='=':
            return print("Tocno")
        else:
            return print("Netocno")
    else:
        if c =='>':
            return print("Tocno")
        elif c=='<':
            return print("Tocno")
        elif c=='=':
            return print("Tocno")
        else:
            return print("Netocno") 
```
